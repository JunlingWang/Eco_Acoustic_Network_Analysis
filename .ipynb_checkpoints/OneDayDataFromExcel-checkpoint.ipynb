```python
import xlrd # This module helps read data from Excel file
import networkx as nx
import numpy as np
import pandas as pd
%matplotlib notebook

# read data from Excel file from local addres
loc = ("Se 2013 Oct Cluster-node list.xlsx")
wb = xlrd.open_workbook(loc)

def getEdge(startMinute):
    sheet = wb.sheet_by_index(0) #the index of sheet
    startVertice = sheet.cell_value(startMinute-1, 2) #(row, column) Python starts a list with 0, hence the need of -1
    endVertice = sheet.cell_value(startMinute, 2) 
    return (int(startVertice), int(endVertice)) # Return the edge. Convert numbers to int, which looks better

# Instantiate the graph
G1 = nx.MultiDiGraph() # Allow self-loop

# add node/edge pairs
i = 0
while i < 1435: # exhaust the whole spreadsheet
    edge = getEdge(i)
    G1.add_edge(edge[0], edge[1])
    i += 1
    
    
#draw the network G1
#nx.draw_networkx(G1)
nx.draw_circular(G1)
G1.nodes_with_selfloops()
```
